```python
#As you can see, most of the HDIForYear value is empty. That's why I want this value deleted.
data=data.drop(['HDIForYear','CountryYear'],axis=1)
```

```python
suicidesNo=[]
for country in data.Country.unique():
    suicidesNo.append(sum(data[data['Country']==country].suicides_no))   
```

```python
suicidesNo=pd.DataFrame(suicidesNo,columns=['suicides_no'])
country=pd.DataFrame(data.Country.unique(),columns=['country'])
data_suicide_countr=pd.concat([suicidesNo,country],axis=1)
#sns.barplot(x=data.Country.unique(),y=suicidesNo) 
#plt.show()
```

```python
data_suicide_countr=data_suicide_countr.sort_values(by='suicides_no',ascending=False)
```

```python
sns.barplot(y=data_suicide_countr.country[:10],x=data_suicide_countr.suicides_no[:10])
plt.show()
```

```python
data_suicide_countr2=data_suicide_countr.sort_values(by='suicides_no',ascending=True)
sns.barplot(y=data_suicide_countr2.country[:10],x=data_suicide_countr2.suicides_no[:10])
plt.show()
```
